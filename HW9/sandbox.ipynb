```python
import spacy
import pandas as pd
```

```python
df = pd.read_csv('data.csv', encoding='utf-8')
```

```python
df
```

```python
nlp = spacy.load("ru_core_news_md")
```

```python
words = list((df['Слово_']))
```

```python
lemmas = []
for i in words:
    i = nlp(i)
    i = i[0]
    lemmas.append(i.lemma_)
```

```python
pos = []
for i in words:
    i = nlp(i)
    i = i[0]
    pos.append(i.pos_)
```

```python
df['Часть речи'] = pos
```

```python
df['Лемма слова'] = lemmas
```

```python
df
```

```python
df.to_csv('data_with_lemmas.csv', encoding='utf-8')
```
